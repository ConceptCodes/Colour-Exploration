# Colour Exploration

This is more of gist type project, I did some simple colour exploration using matplotlib, plotly and pandas. The dataset is from wikipedia article on a list of colours from a-z, I used a web scraper to get the data and stored it as a csv file for later extraction.
```python
#import libiaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
#extract data and remove unwanted columns
colours = pd.DataFrame(pd.read_csv('colours.csv'))
#we use axis=1 for columns, axis=0 would be for rows
colours = colours.drop(['Hue','Satur','Light','Satur.1','Value'], axis=1)
print('There are {} colours in this dataset sourced from wikipedia'.format(len(colours)))
colours.head()
```

```python
#convert the string of the number to a float and divide by 255 to ge the value in range of 0 -1
#if confused on how i did this look up list operations
def normalize(arr):
    return [(float(i.replace('%','')) / 255.0) for i in arr]
r = normalize(colours['Red'])
g = normalize(colours['Green'])
b = normalize(colours['Blue'])
```

```python
#plot the colours in 3D space and assign the colour to it
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(r, g, b, color=colours['Hex'])
plt.show()
```
